```python
# Filling null Values with a  pervious value
df3 = df.fillna(method = 'pad')
df3
```

```python
# Filling null value with the next value 
df4 = df.fillna(method = 'bfill')
df4
```

```python
# Filling Null value with the mean of a column
# df1 = df.copy()
df1 = df.fillna(value = df["Age"].median())
df1 = df.fillna(value = df["Fare"].median())
```

```python
df1.isnull().sum()
```

```python
df1.hist(bins=20, figsize=(15, 10))
plt.show()
```

```python
sns.countplot(x='Pclass', data=df1)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
# Male is more Than Female
sns.countplot(x='Sex', data=df1)
plt.title('Gender Distribution')
plt.show()
```

```python
sns.countplot(x='Embarked', data=df1)
plt.title('Embarked Distribution')
plt.show()
```

```python
# The relations between the variables
sns.scatterplot(x='Age', y='Fare', data=df1, hue='Pclass' )
plt.title('Age vs Fare with Pclass')
plt.show()
```

```python
# The relations between the variables
sns.scatterplot(x='Sex', y='Age', data=df1, hue='Pclass' )
plt.title('Sex vs Age with Pclass')
plt.show()
```
